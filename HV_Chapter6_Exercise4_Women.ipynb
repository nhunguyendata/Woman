```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wis=pd.read_csv('women_in_science.csv',index_col=0)
print(wis.shape)
print(wis.info())
wis
```

```python
wis.shape
```

```python
# Câu 1: Vẽ biểu đồ line plot thể hiện tỷ lệ % bằng Khoa học vật lý và 
# Khoa học máy tính được trao cho phụ nữ qua các năm
# Đồ thị có 2 line:
# line 1: year, physical_sciences, color='blue'; và  
# line 2: year,computer_science, color='red'
# Bạn nhận xét gì về biểu đồ vừa vẽ
plt.plot(wis['year'],wis['physical_sciences'],color='r',label='Physical_sciences')
plt.plot(wis['year'],wis['computer_science'],color='g',label='Computer_science')
plt.legend()
plt.show()
```

```python
# Câu 4: Vẽ 4 biểu đồ line với 4 màu khác nhau: 
# Biểu đồ 1: thể hiện tỷ lệ % bằng Khoa học vật lý 
# được trao cho phụ nữ qua các năm (year - physical_sciences) 
# Biểu đồ 2: thể hiện tỷ lệ % bằng Khoa học máy tính 
# được trao cho phụ nữ qua các năm (year - computer_science)
# Biểu đồ 3: thể hiện tỷ lệ %  phụ nữ tham gia 
# các công việc liên quan đến y tế qua các năm (year - health)
# Biểu đồ 4: thể hiện tỷ lệ %  phụ nữ tham gia 
# các công việc liên quan đến giáo dục qua các năm (year - education)

# trên 4 subplot: plt.subplot(2, 2, 1), plt.subplot(2, 2, 2), 
# plt.subplot(2, 2, 3), plt.subplot(2, 2, 4)
# Bạn nhận xét gì về biểu đồ vừa vẽ
plt.subplot(2,2,1)
plt.plot(wis['year'],wis['physical_sciences'], color='green')
plt.title("Physical Science")
plt.subplot(2,2,2)
plt.plot(wis['year'],wis['computer_science'], color='red')
plt.title("Computer Science")
plt.subplot(2,2,3)
plt.plot(wis['year'],wis['health'], color='yellow')
plt.title("Health")
plt.subplot(2,2,4)
plt.plot(wis['year'],wis['education'], color='purple')
plt.title("Education")
plt.tight_layout()
plt.show()
```

```python
# Câu 5: Vẽ 2 biểu đồ line plot: 
# Biểu đồ 1: thể hiện tỷ lệ % bằng Khoa học máy tính 
# được trao cho phụ nữ qua các năm (year - computer_science)
# Biểu đồ 2: thể hiện tỷ lệ % bằng Khoa học vật lý 
# được trao cho phụ nữ qua các năm (year - physical_sciences) 
# nhưng giới hạn từ năm 1980-2000 --> có plt.xlim(1980, 2000) và plt.ylim(0, 50) 
plt.plot(wis['year'],wis['physical_sciences'], color='green',label='Physical sciences')
plt.plot(wis['year'],wis['computer_science'], color='red',label='Computer science')
plt.xlabel('Năm')
plt.ylabel('Phần trăm')
plt.title('Tỷ lệ % phụ nữ được trao giải từ 1980-2000')
plt.xlim(1980, 2000)
plt.ylim(0, 50)
plt.xticks(np.arange(1980, 2004, 3))
plt.legend()
plt.tight_layout()
plt.show()
```
